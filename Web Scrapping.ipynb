```python
import bs4
import requests
import pandas as pd

res=requests.get('https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off')
type(res)
soup=bs4.BeautifulSoup(res.text,"lxml")
name=soup.select("._3wU53n")
price=soup.select("._1vC4OE")
names = []
prices = []
for i in range(24):
    names.append(name[i].text)
    prices.append(price[i].text)
List=pd.DataFrame({'product':names,'links':prices})
List.to_excel('data.xls')
List
```
